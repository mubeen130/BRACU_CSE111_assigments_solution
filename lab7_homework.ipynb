```python
#task1
import math
class Circle:
    def __init__(self,radius):
        self.__radius = radius
    def getRadius(self):
        return self.__radius
    def setRadius(self,radius1):
        self.__radius = radius1
    def area(self):
        val = math.pi*(self.__radius)**2
        return val

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
#task2
class Triangle:
    def __init__(self,base,height):
        self.__base = base
        self.__height = height

    def getBase(self):
        return self.__base
    def getHeight(self):
        return self.__height

    def setHeight(self,a):
        self.__height = a
    def setBase(self,b):
        self.__base = b
    def area(self):
        val = .5*self.__base*self.__height
        return val


t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#task3
class Team:
    def __init__(self,name = 0):
        self.__name = name
        self.__li = []
    def setName(self,nm):
        self.__name = nm
    def addPlayer(self,obj):
        self.__li.append(obj.player_name)
    def printDetail(self):
        print(f"Team: {self.__name}")
        print(f"List of players:\n{self.__li}")
class Player:
    def __init__(self,player_name):
        self.player_name = player_name

b = Team()
b.setName('Bangladesh')
mashrafi = Player("Mashrafi")
b.addPlayer(mashrafi)
tamim = Player("Tamim")
b.addPlayer(tamim)
b.printDetail()
a = Team("Australia")
ponting = Player("Ponting")
a.addPlayer(ponting)
lee = Player("Lee")
a.addPlayer(lee)
a.printDetail()
```

```python
#task4
class Spaceship:
    def __init__(self,name,total_c):
        self.name = name
        self.total_c = total_c
        self.li = []
        self.val = 0
    def load_cargo(self,obj):
        self.val += obj.getw()
        if self.val<=self.total_c:
            self.li.append(obj.getName())
        else:
            print(f"Warning: Unable to load {obj.getName()} inside {self.name}. Exceeds capacity by {(self.val-self.total_c)}.")
    def display_details(self):
        print(f"Spaceship Name: {self.name}")
        print(f"Capacity: {self.total_c}")
        print(f"Current Cargo Weight: {self.val}")
        print(f"Cargo: {self.li}")

class Cargo:
    def __init__(self,name,tw):
        self.__name = name
        self.__tw = tw
    def getName(self):
        return self.__name
    def getw(self):
        return self.__tw

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#task5
class BracuStudent:
    def __init__(self,name,home):
        self.name = name
        self.home = home
        self.count = False
    def get_pass(self):
        self.count = True
    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.count}")
class BracuBus:
    def __init__(self,bus_name,seat = 2):
        self.bus_name = bus_name
        self.seat = seat
        self.li = []
    def board(self,*args):
        self.count2 = 0
        if len(args) == 0:
            print("No passengers!")
        else:
            for i in args:
                if i.home != self.bus_name:
                    print("You got on the wrong bus!")
                elif i.count == False:
                    print("You don't have a bus pass!")
                else:
                    if self.count2<self.seat:
                        print(f"{i.name} boarded the bus.")
                        self.count2+=1
                        self.li.append(i.name)
                    else:
                        print("Bus is full!")
    def show_details(self):
        print(f"Bus Route: {self.bus_name}")
        print(f"Passengers Count: {len(self.li)} Max:({self.seat})")
        print(f"Passengers On Board: {self.li}")




st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur",5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1,st2,st3)
print("10===========================")
bus1.show_details()
```

```python
#task 6
class Library:
    def __init__(self,name,dict1):
        self.name = name
        self.dict1 = dict1
        self.dict2 = {}
    def details(self):
        print(f"{self.name} Library details")
        print("Borrower details:")
        print(self.dict2)
        print(f"Books availability:\n{self.dict1}")
class Reader:
    def __init__(self,name2):
        self.name2 = name2
        self.dict3 = {}
        self.count = 0
    def borrow(self,obj,*args):
        for i in args:
            if self.count <5:
                if obj.dict1[i] != 0:
                    self.count += 1
                    if i not in self.dict3.keys():
                        self.dict3[i] = 1
                        obj.dict1[i] -= 1
                    else:
                        self.dict3[i] += 1
                        obj.dict1[i] -= 1
                    print(f"{i} book is borrowed successfully.")
                else:
                    print(f"{i} books are not available at the moment.")
            else:
                print("You cannot borrow more than 5 books.")
            obj.dict2[self.name2] = self.count
    def readerInfo(self,val=None):
        if val == None:
            print(f"{self.name2}, you have {self.count} book(s) with you.")
            for i,j in self.dict3.items():
                print(f"Books on {i}:{j}")
        else:
            print(f"{self.name2} you have {self.dict3[val]} {val} book(s) with you")


L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
