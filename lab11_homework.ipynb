```python
#task 1
class Employee:
  def __init__(self, name, id, dept):
    self.name = name
    self.id = id
    self.dept = dept
    self.type = "Regular"
    self.taka = 30000
  def workDistribution(self, dept2):
    temp = ""
    if dept2 == "Human Resource":
      temp += "Collect work distribution details from the manager."
    else:
      temp += "Collect work distribution loads from the HR department."
    print(temp)
  def increment(self):
    if self.type == "Foreign":
      self.taka += ((15/100)*self.taka)
    elif self.type == "Part Time":
      print("Sadly, there is no increment for the part time employees!!")
    else:
      self.taka += (10/100)*self.taka
  def employeeDetails(self):
    temp2 = f"Name: {self.name}, Dept {self.dept}\nEmployee id: {self.id}, Salary: {self.taka}"
    if self.type == "Regular":
      pass
    else:
      temp2 += f"\nEmployee Type: {self.type}"
    print(temp2)
class Foreign_employee(Employee):
  def __init__(self, name, id, dept):
    super().__init__(name, id, dept)
    self.type = "Foreign"
class Part_time_employee(Employee):
  def __init__(self, name, id, dept):
    super().__init__(name, id, dept)
    self.type = "Part Time"
    self.taka = 15000

print("1------------------------------------------")
emp1=Employee("Nawaz Ali", 102, "Marketing")
print("2------------------------------------------")
emp1.employeeDetails()
print("3------------------------------------------")
emp1.workDistribution("Marketing")
print("4------------------------------------------")
emp1.increment()
emp1.employeeDetails()
print("5------------------------------------------")
f_emp=Foreign_employee("Nadvi", 311, "Human Resource")
f_emp.employeeDetails()
print("6------------------------------------------")
f_emp.workDistribution("Human Resource")
print("7------------------------------------------")
f_emp.increment()
f_emp.employeeDetails()
print("8------------------------------------------")
p1_emp=Part_time_employee("Asif", 210, "Sales")
p2_emp=Part_time_employee("Olive", 223, "Accounts")
print("9------------------------------------------")
p1_emp.employeeDetails()
print("10------------------------------------------")
p1_emp.workDistribution("Sales")
print("11------------------------------------------")
p2_emp.increment()
print("12------------------------------------------")
p2_emp.employeeDetails()
```

```python
#task 2
class Exam:
    def __init__(self, marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
class ScienceExam(Exam):
    def __init__(self,marks, time=0,*subs):
        super().__init__(marks)
        self.time = time
        self.subs = subs
        self.li2 = []
        self.li = []
        self.li = super().examSyllabus().split(" , ")
        self.li2 += self.li + [i for i in subs]
    def __str__(self):
        return f"Marks : {self.marks} Time: {self.time} Number of parts: {len(self.subs)+len(self.li)}"
    def examSyllabus(self):
        temp = ""
        for j in range(len(self.li2)):
            temp+= self.li2[j]
            if j < len(self.li2)-1:
                temp+=" , "
        return temp
    def examParts(self):
        temp2 = ""
        temp2 += super().examParts()
        count = len(super().examSyllabus().split(" , "))
        for m in range(len(self.subs)):
            count+=1
            if m == len(self.subs)-1:
                temp2+= f"Part {count} - {self.subs[m]}"
            else:
                temp2 += f"Part {count} - {self.subs[m]}\n"
        return temp2


engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

```

```python
#task 3
class PokemonBasic:
  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type
  def get_type(self):
    return 'Main type: ' + self.type
  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'
  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness
class PokemonExtra(PokemonBasic):
    def __init__(self,name = 'Default', hp = 0, weakness = 'None', type = 'Unknown',sec_type = None,other = None):
        super().__init__(name,hp,weakness,type)
        self.sec_type = sec_type
        self.other = other
    def get_type(self):
        temp = super().get_type()
        if self.sec_type != None:
            temp += f", Secondary type: {self.sec_type}"
        return temp
    def get_move(self):
        temp2 = super().get_move()
        if self.other != None:
            temp2 += f"\nOther move: "
            for m in range(len(self.other)):
                if m == len(self.other) - 1:
                    temp2 += f"{self.other[m]}"
                else:
                    temp2 += f"{self.other[m]}, "
        return temp2


print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#task 4
class Cakes:
    order_list = {}
    feedback_dict = {}
    def __init__(self, flavor, weight):
        self.flavor = flavor
        self.weight = weight
        self.sweetness = "Moderate sugar"
        self.cream_type = "Whipped Cream"
        self.price = (1200/1000) * self.weight
        key = f"{self.flavor} Cake {self.weight}gm"
        Cakes.order_list[key] = 1
    def add_customization(self, sweetness, cream_type):
        self.sweetness = sweetness
        self.cream_type = cream_type
    def cake_details(self):
        print(f"Flavor: {self.flavor} Cake, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness} sugar, Cream Type: {self.cream_type}")
        print(f"Price: {self.price} Taka")
    @classmethod
    def give_feedbacks(cls, cake_fl, comments):
        print(f"Thanks for your valuable feedback!")
        if "Cheese Cake" in cake_fl:
            print("You will get 10% discount for your next purchase!")
        if cake_fl not in cls.feedback_dict:
            cls.feedback_dict[cake_fl] = [comments]
        else:
            cls.feedback_dict[cake_fl].append(comments)

    @classmethod
    def show_feedbacks(cls):
        print(str(cls.feedback_dict))
class Cheese_Cakes(Cakes):
    def __init__(self, flavor, weight, cake_type="baked"):
        super().__init__(flavor, weight)
        self.price = (1500 / 1000) * self.weight
        self.type = cake_type
        self.cream_type = "Cream Cheese"
    def add_customization(self):
        print("Sorry! No customization available for cheese cakes.")
    def cake_details(self):
        temp = ""
        temp += f"Flavor: {self.flavor} Cheese Cake, Weight: {self.weight} gm\n"
        temp += f"Sweetness: {self.sweetness}, Cream Type: {self.cream_type}\n"
        temp += f"Cake Type: {self.type.capitalize()}, Price: {self.price} Taka"
        print(temp)
order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake", "Very Delicious")
Cakes.give_feedbacks("Chocolate Cake", "Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake", "Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake", "Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
