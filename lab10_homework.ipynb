```python
#task 1
class Shape:
  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base
  def get_height_base(self):
    return "Height: "+str(self.height)+", Base: "+str(self.base)
class triangle(Shape):
    def __init__(self,name='Default', height=0, base=0):
        super().__init__(name,height,base)
    def calcArea(self):
        self.area = 0.5*self.height*self.base
    def printDetail(self):
        temp = f"Shape name: {self.name}\n"
        temp+= f"Height: {self.height}, "
        temp += f"Base: {self.base}\n"
        temp += f"Area: {self.area}"
        return temp
class trapezoid(Shape):
    def __init__(self,name='Default', height=0, base=0,width=0):
        super().__init__(name,height,base)
        self.width = width
    def calcArea(self):
        self.area = ((self.base+self.width)/2)*self.height
    def printDetail(self):
        temp = f"Shape name: {self.name}\n"
        temp+= f"Height: {self.height}, "
        temp += f"Base: {self.base}, "
        temp += f"Side_A: {self.width}\n"
        temp += f"Area: {self.area}"
        return temp

#Driver code:
tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
#task 2
class SportsPerson:
  def __init__(self, team_name, name, role):
    self.__team = team_name
    self.__name = name
    self.role = role
    self.earning_per_match = 0
  def get_name_team(self):
    return 'Name: '+self.__name+', Team Name: ' +self.__team
class Player(SportsPerson):
    def __init__(self,team_name, name, role,goal,played):
        super().__init__(team_name, name, role)
        self.goal = goal
        self.played = played
        self.earn = (goal*1000)+(played*10)
    def calculate_ratio(self):
        return self.goal/self.played
    def print_details(self):
        temp = super().get_name_team()+"\n"
        temp += f"Team Role: {self.role}\n"
        temp += f"Team Goal: {self.goal}, Total Played: {self.played}\n"
        temp += f"Goal Ratio: {self.calculate_ratio()}\n"
        temp += f"Match Earning: {self.earn}k"
        print(temp)
class Manager(SportsPerson):
    def __init__(self, team_name, name, role, win):
        super().__init__(team_name, name, role)
        self.win = win
        self.earn = win*1000
    def print_details(self):
        temp = super().get_name_team()+"\n"
        temp += f"Team Role: {self.role}\n"
        temp += f"Total win: {self.win}\n"
        temp += f"Match Earning: {self.earn}k"
        print(temp)


#driver code:
player_one = Player('Al-Nassr', 'Ronaldo', 'Striker', 25, 32)
player_one.calculate_ratio()
player_one.print_details()
print('------------------------------------------')
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25)
manager_one.print_details()
```

```python
#task 3
class KK_tea:
    dict1={}
    def __init__(self,price,nos_bag=50):
        self.name = "KK Regular Tea"
        self.price = price
        self.nos_bag = nos_bag
        self.flag = False
        self.count = 0
        if self.name not in KK_tea.dict1.keys():
            KK_tea.dict1[self.name] = self.count
        else:
            KK_tea.dict1[self.name] += self.count

    def product_detail(self):
        temp = f"Name : {self.name}, Weight: {self.nos_bag*2}\n"
        temp += f"Tea Bags: {self.nos_bag}, Price: {self.price}\n"
        temp += f"Status: {self.flag}"
        print(temp)
    @classmethod
    def total_sales(cls):
        print(f"Total Sales: {cls.dict1}")
    @classmethod
    def update_sold_status_regular(cls,*args):
        for i in args:
            i.flag = True
        KK_tea.dict1[args[1].name] = len(args)
class KK_flavoured_tea(KK_tea):
    def __init__(self,flavour,price,nos_bag=50):
        super().__init__(price,nos_bag)
        self.flavour = flavour
        self.name = f"KK {flavour} tea"
    @classmethod
    def update_sold_status_flavoured(cls,*args):
        li = []
        count2 = 0
        for i in args:
            count2 += 1
            i.flag = True
            li.append(i.name)
        for k in li:
            if k not in cls.dict1.keys():
                cls.dict1[k] = 1
            else:
                cls.dict1[k] += 1


#driver code:
t1 = KK_tea(250)
print("-----------------1-----------------")
t1.product_detail()
print("-----------------2-----------------")
KK_tea.total_sales()
print("-----------------3-----------------")
t2 = KK_tea(470, 100)
t3 = KK_tea(360, 75)
KK_tea.update_sold_status_regular(t1, t2, t3)
print("-----------------4-----------------")
t3.product_detail()
print("-----------------5-----------------")
KK_tea.total_sales()
print("-----------------6-----------------")
t4 = KK_flavoured_tea("Jasmine", 260, 50)
t5 = KK_flavoured_tea("Honey Lemon", 270, 45)
t6 = KK_flavoured_tea("Honey Lemon", 270, 45)
print("-----------------7-----------------")
t4.product_detail()
print("-----------------8-----------------")
t6.product_detail()
print("-----------------9-----------------")
KK_flavoured_tea.update_sold_status_flavoured(t4, t5, t6)
print("-----------------10-----------------")
KK_tea.total_sales()
```

```python
#task 4
class TwoDVector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add2DVectors(self, *vectors):
        for i in vectors:
            self.x += i.x
            self.y += i.y
    def print2DVector(self):
        if self.y >= 0:
            y = "+ "+str(self.y)
        else:
            y = str(self.y)
        print(f"{self.x}i {y}j")
class ThreeDVector(TwoDVector):
    def __init__(self, x, y, z):
        super().__init__( x, y)
        self.z = z
    def add3DVectors(self,*vectors):
        super().add2DVectors(*vectors)
        for i in vectors:
            self.z += i.z
    def multiplyScalar(self,sc):
        self.x *= sc
        self.y *= sc
        self.z *= sc
    def print3DVector(self):
        temp = ""
        if self.x >= 0:
            temp+= f"{self.x}i"
        else:
            temp+= f" - {self.x}i"
        if self.y >= 0:
            temp += f" + {self.y}j"
        else:
            temp += f" {self.y}j"
        if self.z >= 0:
            temp += f" + {self.z}k"
        else:
            temp += f" {self.z}k"
        print(temp)
    def calculateLength(self):
        return (self.x**2 + self.y**2 + self.z**2)**0.5


#driver code:
TwoDV1 = TwoDVector(5, 6)
TwoDV2 = TwoDVector(3, 7)
TwoDV3 = TwoDVector(1, 8)
print("===============")
TwoDV1.add2DVectors(TwoDV2, TwoDV3)
TwoDV1.print2DVector()
print("===============")
ThreeDV1 = ThreeDVector(5, 6, 1)
ThreeDV2 = ThreeDVector(1, 9, -7)
ThreeDV3 = ThreeDVector(8, 2, 4)
print("===============")
ThreeDV1.add3DVectors(ThreeDV2,ThreeDV3)
ThreeDV1.print3DVector()
print("===============")
ThreeDV1.multiplyScalar(3)
ThreeDV1.print3DVector()
print("===============")
print(ThreeDV1.calculateLength())
```
