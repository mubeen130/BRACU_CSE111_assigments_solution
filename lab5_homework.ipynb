```python
#Task 1
class Student:
    def __init__(self,name,cgpa,credit=9,dept="CSE"):
        self.name = name
        self.cgpa = cgpa
        self.credit = credit
        self.dept = dept
        self.count = ""
    def checkScholarshipEligibility(self):
        if self.cgpa>=3.5 and self.credit>10:
            if self.cgpa>=3.7:
                print(f"{self.name} is eligible for merit Based scholarship")
                self.count = "Merit-based Scholarship"
            elif 3.7>self.cgpa>=3.5:
                print(f"{self.name} is eligible for need based scholarship")
                self.count = "Need-Based Scholarship"
        else:
            print(f"{self.name} is not eligible for scholarship")
            self.count = "No Scholarship"
    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Dept: {self.dept}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of credits: {self.credit}")
        print(f"Scholarship status: {self.count}")


print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#task 2
menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}
class Foodie:
    def __init__(self,name):
        self.name = name
        self.li = []
        self.sum = 0
    def order(self,*args):
        dict1 = {}
        for i in args:
            val1,val2 = i.split("-")
            dict1[val1] = val2
            self.li.append(val1)
        for m,n in dict1.items():
            print(f"Ordered - {m}, quantity - {n}, price(per unit) - {menu[m]}.")
            print(f"Total price: {menu[m]*int(n)}")
            self.sum+= menu[m]*int(n)
    def pay_tips(self,tip = 0):
        self.tip = tip
        if self.tip == 0:
            print("No tips to the waiter.")
        else:
            print(f"Gives {self.tip}/- tips to the waiter.")
            self.sum+=self.tip
    def show_orders(self):
        return f"{self.name} has {len(self.li)} item(s) in the cart.\n" \
               f"Items: {self.li}\nTotal spent: {self.sum}"

f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#task3
class Department:
    def __init__(self,dept="ChE",val1 = 5):
        self.val1 = val1
        self.dept = dept
        print(f"The {self.dept} has {self.val1} sections")
    def add_students(self,*nums):
        self.li = []
        self.sum = 0
        if len(nums) == self.val1:
            for j in nums:
                self.li.append(j)
                self.sum+=int(j)
            avg = round(self.sum/len(nums),2)
            print(f"The {self.dept} has an average of {avg} students in each section")
        elif len(nums) != self.val1:
            print(f"{self.dept} doesnt have {len(nums)} sections")
    def merge_Department(self,*args):
        for i in args:
            self.sum+=i.sum
            print(f"{i.dept} is merged to {self.dept}.")
        avg1 = self.sum/len(self.li)
        return f"Now the {self.dept} has an average of {round(avg1,2)} students in each section."


d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#task 4
class Shopidify:
    def __init__(self,name="Guest"):
        self.name = name
        if self.name == "Guest":
            print("Welcome to shopidify")
        else:
            print(f"Welcome {self.name} to shopidify")
        self.dict1 = {}
        self.count = 0
    def add_to_cart(self, goods, val=1):
        self.count2 = 0
        list1 = []
        list2 = []
        self.val = val
        self.count2 += val
        if type(goods) == str:
            if goods in self.dict1:
                self.dict1[goods] += self.count2
            else:
                self.dict1[goods] = self.count2
        else:
            for i in goods:
                if type(i) == str:
                    list1.append(i)
                else:
                    list2.append(i)
            for m in range(len(list1)):
                for n in range(len(list2)):
                    if m == n:
                        if list1[m] in self.dict1:
                            self.dict1[list1[m]] += list2[n]
                        else:
                            self.dict1[list1[m]] = list2[n]
    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for i,j in self.dict1.items():
            print(f"- {i} : {j}x")
    def checkout(self):
        self.count+=1
        print(f"Checkout completed for {self.name}")
    def display_history(self):
        print(f"Purchase history for {self.name}:")
        print(f"Transaction {self.count}")
        for i,j in self.dict1.items():
            print(f"- {i} : {j}x")

guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
#task5
class Author:
    def __init__(self,name = 0):
        self.name = name
        self.count = 0
        self.dict1 = {}
    def setName(self,name2):
        self.name = name2
    def addBook(self,bookName,bookType):
        if self.name == 0 :
            print("A book can not be added without author name")
        else:
            self.bookName = bookName
            self.bookType = bookType
            self.count+=1
            if self.bookType in self.dict1.keys():
                self.dict1[self.bookType] += f",{self.bookName}"
            else:
                self.dict1[self.bookType] = self.bookName
    def printDetail(self):
        print("Number of books:", self.count)
        print("Author name:",self.name)
        for i,j in self.dict1.items():
            print(f"{i}:{j}")

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
